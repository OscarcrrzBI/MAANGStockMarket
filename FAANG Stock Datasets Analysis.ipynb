```python
amazon = pd.read_csv("downloads/AMAZON_daily.csv")
apple = pd.read_csv("downloads/APPLE_daily.csv")
google = pd.read_csv("downloads/GOOGLE_daily.csv")
meta = pd.read_csv("downloads/META_daily.csv")
netflix = pd.read_csv("downloads/NETFLIX_daily.csv")
```

```python
#Create a new dataframe with all files merge together
dfs = [amazon, apple, google, meta, netflix]
```

```python
#Create two new columns that get the rolling average of the firts 50 and the firts 200 close values
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
#Check new column values starting on 50
amazon.head(50)
```

```python
#Starting on 200 mean
amazon.head(200)
```

```python
#Create a new column to show the last close price by shifting 1 the CLose column
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
amazon.head()
```

```python
#Add new column 
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
amazon.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
amazon.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
amazon.head()
```

```python
for df in dfs:
    df['Change in Volume'] = df['Volume'] - df['Previous day volume']  
```

```python
amazon.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
